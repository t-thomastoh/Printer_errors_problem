```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Barchart

```python
open_file = '../Python_practice/data_source/countries.csv'
countries_file = pd.read_csv(open_file)
```

```python
plt.figure(figsize=(10,7))
plt.title('World population growth rate')
sns.barplot(x=countries_file['year'], y=countries_file['population'])
```

```python
plt.figure(figsize=(10,7))
plt.title('Singapore population growth rate')
sns.barplot(x=countries_file['year'][1356:1368], y=countries_file['population'][1356:1368])
```

# Heatmap

```python
open_file = '../Python_practice/data_source/flight_delay.csv'
flight_delay_file = pd.read_csv(open_file)
```

```python
plt.figure(figsize=(14,7))
plt.title("Flight delay")
sns.heatmap(data=flight_delay_file, annot=True) #'annot' provide the number to indicate the delay time
```

Note: Heatmap seems to encounter a bit of problem when the data is too large. Doesn't seems like the right chart to use for big data too.

# Line Charts

```python
open_file = '../Python_practice/data_source/spotify.csv'
spotify_file = pd.read_csv(open_file, index_col="Date", parse_dates=True)
```

# Distributions

```python
flower_file = '../Python_practice/data_source/iris.csv'
flower_data = pd.read_csv(flower_file)
```

```python
sns.distplot(a=flower_data['Petal Length (cm)'], kde=False)
```

```python
sns.kdeplot(data=flower_data['Petal Length (cm)'], shade=True)
```

```python
sns.jointplot(x=flower_data['Petal Length (cm)'], y=flower_data['Sepal Width (cm)'], kind="kde")
```

```python
iris_set_filepath = '../Python_practice/data_source/iris_setosa.csv'
iris_ver_filepath = '../Python_practice/data_source/iris_versicolor.csv'
iris_vir_filepath = '../Python_practice/data_source/iris_virginica.csv'

iris_set_data = pd.read_csv(iris_set_filepath, index_col="Id")
iris_ver_data = pd.read_csv(iris_ver_filepath, index_col="Id")
iris_vir_data = pd.read_csv(iris_vir_filepath, index_col="Id")
```

```python
sns.distplot(a=iris_set_data['Petal Length (cm)'], label="Iris-setosa", kde=False)
sns.distplot(a=iris_ver_data['Petal Length (cm)'], label="Iris-versicolor", kde=False)
sns.distplot(a=iris_vir_data['Petal Length (cm)'], label="Iris-virginica", kde=False)

plt.title("Histogram of Petal Lengths, by Species")

# Force legend to appear
plt.legend()
```

```python
# KDE plots for each species
sns.kdeplot(data=iris_set_data['Petal Length (cm)'], label="Iris-setosa", shade=True)
sns.kdeplot(data=iris_ver_data['Petal Length (cm)'], label="Iris-versicolor", shade=True)
sns.kdeplot(data=iris_vir_data['Petal Length (cm)'], label="Iris-virginica", shade=True)

# Add title
plt.title("Distribution of Petal Lengths, by Species")
```
