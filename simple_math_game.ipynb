## Simple Math Game

1. Participants is able to choose how many questions they woud like to answer per level.
2. Equation is randomly generated and so is the operator

```python
import random

def math_ques(num_of_qns):
    questions_level1 = {}
    score_level1 = 0
    total_qns_level1 = 0
    
    for i in range(int(num_of_qns)):
        int_1 = random.randint(0,10) #Able to change the index?
        int_2 = random.randint(0,10)
        operators = ['+','-'] #operator can be arrange an universal varaible rather than local
        operator_value = random.choice(operators)
        question = str(int_1) + ' ' + str(operator_value) + ' ' + str(int_2)
        answer = str(eval(question))
        question+=": "
    
        questions_level1.update({question:answer})

    for q in questions_level1.keys():
        user_answer = input(q)
        if questions_level1.get(q) == str(user_answer):
            score_level1 += 1
            total_qns_level1 += 1
            print('Correct!')
        else:
            print('Incorrect! Your answer is ' + str(user_answer) + ', where ' + questions_level1.get(q) + ' is the correct answer.')
            total_qns_level1 += 1
    print('Level 1: ' + str(score_level1) + '/' + str(total_qns_level1))
    user_choice = input("Press Enter to proceed to level 2 or Q to quit: ")
    if user_choice.capitalize() == 'Q':
        print('Level 1: ' + str(score_level1) + '/' + str(total_qns_level1))
        return
    
    questions_level2 = {}
    score_level2 = 0
    total_qns_level2 = 0

    for i in range(int(num_of_qns)):
        int_1 = random.randint(10,100)
        int_2 = random.randint(10,100)
        operators = ['+','-']
        operator_value = random.choice(operators)
        question = str(int_1) + ' ' + str(operator_value) + ' ' + str(int_2)
        answer = str(eval(question))
        question+=": "
    
        questions_level2.update({question:answer})

    for q in questions_level2.keys():
        user_answer = input(q)
        if questions_level2.get(q) == str(user_answer):
            score_level2 += 1
            total_qns_level2 += 1
            print('Correct!')
        else:
            print('Incorrect! Your answer is ' + str(user_answer) + ', where ' + questions_level2.get(q) + ' is the correct answer.')
            total_qns_level2 += 1
    print('Level 2: ' + str(score_level2) + '/' + str(total_qns_level2))
    user_choice = input("Press Enter to proceed to level 3 or Q to quit: ")
    if user_choice.capitalize() == 'Q':
        print('Level 1: ' + str(score_level1) + '/' + str(total_qns_level1))
        print('Level 2: ' + str(score_level2) + '/' + str(total_qns_level2))
        return
    
    questions_level3 = {}
    score_level3 = 0
    total_qns_level3 = 0

    for i in range(int(num_of_qns)):
        int_1 = random.randint(100,1000)
        int_2 = random.randint(100,1000)
        operators = ['+','-']
        operator_value = random.choice(operators)
        question = str(int_1) + ' ' + str(operator_value) + ' ' + str(int_2)
        answer = str(eval(question))
        question+=": "
    
        questions_level3.update({question:answer})

    for q in questions_level3.keys():
        user_answer = input(q)
        if questions_level3.get(q) == str(user_answer):
            score_level3 += 1
            total_qns_level3 += 1
            print('Correct!')
        else:
            print('Incorrect! Your answer is ' + str(user_answer) + ', where ' + questions_level3.get(q) + ' is the correct answer.')
            total_qns_level3 += 1
    print('Level 3: ' + str(score_level3) + '/' + str(total_qns_level3))
    user_choice = input("You have completed! Press Q to quit: ")
    if user_choice.capitalize() == 'Q':
        print('Level 1: ' + str(score_level1) + '/' + str(total_qns_level1))
        print('Level 2: ' + str(score_level2) + '/' + str(total_qns_level2))
        print('Level 3: ' + str(score_level3) + '/' + str(total_qns_level3))
        return
        
math_ques(input('Enter number of question: '))
```
